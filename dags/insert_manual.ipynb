```python
import pandas as pd
from elasticsearch import Elasticsearch

es = Elasticsearch("http://localhost:9200")
print(es.ping())

df = pd.read_csv('P2M3_talitha_salsabila_data_clean.csv')
for i, r in df.iterrows():
    doc = r.to_json()
    res = es.index(index="new_m3_talitha", doc_type="doc", body=doc)
    print(res)
```
